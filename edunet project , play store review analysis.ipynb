## **Data cleaning and Data framimg**



```python
# convert reviews type as integer type
df = df.astype({"Reviews": int})
```

```python
#drop $ sign from price column
df['Price'] = df['Price'].str.replace('$','')
```

```python
#convert price column type as float
df = df.astype({"Price": float})
```

```python
#in installs column we replace some sign and convert it into integer data type
df['Installs'] = df['Installs'].astype(str).str.replace(',','') # Convert to string before replacing
df['Installs'] = df['Installs'].str.replace('+','')
df['Installs'] = df['Installs'].astype(int) # Convert to integer after string operations
```

```python
#In size column apps size in mb and kb
df["Size"].unique()
```

```python
#convert all apps size in mb data and drop some sign for coverting it into float data type
df['Size'] = df['Size'].apply(lambda x: str(x).replace('Varies with device', '20') if 'Varies with device' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: str(x).replace(',', '') if 'M' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: float(x))
```

```python
#After converted
df["Size"].unique()
```

```python
#data info after cleaning
df.info()
```

```python
df.tail()
```

```python
#unique apps
df_df=df["App"].nunique()
df_df
```

```python
#drop the duplicate apps
right_df=df.drop_duplicates(subset='App')
right_df.head()
```

```python
#null values in rating-1463
right_df.isnull().sum()
```
